```python
# Load the Cardio Dataset
mydata = pd.read_csv('watches(cleaned).csv')
```

```python
# shows the first 5 rows of the dataset
# mydata.head()

# shows the column names in your dataset.
# mydata.columns

# The number of rows and columns, column names, data type of each column, How many non-null values are in each column.
mydata.info()
```

```python
# This command gives you summary statistics for all the numeric columns in your dataset.
mydata.describe()
```

```python
import matplotlib.pyplot as plt   # A library used for creating graphs and charts.
# It makes your charts appear directly inside the notebook
%matplotlib inline

mydata.hist(figsize=(30,40))
plt.show()
```

```python
import seaborn as sns

# This creates a box plot that compares the age distribution of customers, grouped by gender.
sns.boxplot(x="Price", y="Movement", data=mydata)
plt.show()
```

```python
# This creates a cross-tabulation table — also called a contingency table — that shows how two categorical variables relate.
pd.crosstab(mydata['Brand'],mydata['Movement'] )
```

```python
pd.crosstab(mydata['Movement'],mydata['Availability'] )
```

```python
# This countplot is a quick visual version of the pd.crosstab() table
sns.countplot(x="Movement", hue="Price", data=mydata)
plt.show()
```

```python
# Group and count how marital status varies across product and gender.
pd.pivot_table(mydata, index=['Brand', 'Gender'],
                     columns=[ 'Movement'], aggfunc=len)
```

```python
# Shows actual values — specifically, the average Income.
pd.pivot_table(mydata,'Price', index=['Brand', 'Gender'],
                     columns=[ 'Shape'])
```

```python
mydata['Price'].std()
```

```python
mydata['Price'].mean()
```

```python
# Creates a histogram that shows the distribution of ages
sns.displot(mydata['Price'])
plt.show()
```

```python
# Tells Pandas to: Take the Age column, Group the data by Gender, Draw a separate histogram for each gender.
mydata.hist(by='Gender',column = 'Price')
’
```

```python
# Tells Pandas to: Take the Age Income, Group the data by Gender, Draw a separate histogram for each gender.
mydata.hist(by='Gender',column = 'Year of production')
plt.show()
```

```python
# Tells Pandas to: Take the Age Miles, Group the data by Gender, Draw a separate histogram for each gender.
mydata.hist(by='Availability',column = 'Shape')
plt.show()
```

```python
mydata.hist(by='Brand',column = 'Condition', figsize=(20,30))
plt.show()
```

```python
# Creates a grid of scatter plots between all numeric columns in your dataset, Each cell shows how two variables are related.
sns.pairplot(mydata)
```
